```python
#HYPOTHESE 2

import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt

#datensatz
file_path = '/content/Datensatz1_v2.csv'
data = pd.read_csv(file_path, delimiter=';')
data['datetime'] = pd.to_datetime(data['Date'] + ' ' + data['timestamp'], format='%d.%m.%y %H:%M')
user1_data = data[data['User_ID'] == 1]
user2_data = data[data['User_ID'] == 2]


def cal_div(data, time_column, content_column): #gruppierung nach datum
    dd = data.groupby(data[time_column].dt.date)[content_column].nunique().reset_index()
    dd.columns = ['date', 'unique_content_count']
    return dd

# 1 und  2
user1_diversity = cal_div(user1_data, 'datetime', 'content')
user2_diversity = cal_div(user2_data, 'datetime', 'content')


ad = pd.merge(user1_diversity, user2_diversity, on='date', suffixes=('_user1', '_user2'))

#gepaarter t-test
t_stat, p_val = stats.ttest_rel(ad['unique_content_count_user1'], ad['unique_content_count_user2'])

plt.figure(figsize=(10, 6))
plt.plot(ad['date'], ad['unique_content_count_user1'], label='User 1', color='purple')
plt.plot(ad['date'], ad['unique_content_count_user2'], label='User 2', color='pink')
plt.xlabel('Datum')
plt.ylabel('Unique Anzahl der Inhalte')
plt.title('Tägliche Unique Anzahl der Inhalte für User 1 und User 2')
plt.xticks(rotation=45)
plt.legend()
plt.show()

#ergebnisse
results = {
    "Paired t-test results for daily diversity of content": {
        "t-statistic": t_stat,
        "p-value": p_val
    }
}

results, ad.head()
```

```python
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt

file_path = '/content/Datensatz1_v2.csv'
data = pd.read_csv(file_path, delimiter=';')
data['datetime'] = pd.to_datetime(data['Date'] + ' ' + data['timestamp'], format='%d.%m.%y %H:%M')
user1_data = data[data['User_ID'] == 1]
user2_data = data[data['User_ID'] == 2]


def calc(data, time_column, content_column, platform_column):  #gruppierung der daten nachh zeit, anzhal der inhalte
    dpl_d = data.groupby([data[time_column].dt.date, platform_column])[content_column].nunique().unstack().fillna(0).reset_index()
    dpl_d.columns.name = None
    return dpl_d

#user 1 und user 2
user1 = calc(user1_data, 'datetime', 'content', 'platform')
user2 = calc(user2_data, 'datetime', 'content', 'platform')

aligned_diversity_platform = pd.merge(user1, user2, on='datetime', suffixes=('_user1', '_user2'))

#gepaarter t-test für jede plattform
platforms = user1.columns[1:]
t_test_r = {}
for platform in platforms:
    t_stat, p_val = stats.ttest_rel(aligned_diversity_platform[f'{platform}_user1'], aligned_diversity_platform[f'{platform}_user2'])
    t_test_r[platform] = {"t-statistic": t_stat, "p-value": p_val}

#plot
fig, axs = plt.subplots(len(platforms), 1, figsize=(14, 12), sharex=True)

for i, platform in enumerate(platforms): #wiederholung der erstellung der tabellen für jede plattform
    axs[i].set_title(f'Tägliche Unique Anzahl der Inhalte für {platform}')
    axs[i].plot(aligned_diversity_platform['datetime'], aligned_diversity_platform[f'{platform}_user1'], label='User 1', color='purple')
    axs[i].plot(aligned_diversity_platform['datetime'], aligned_diversity_platform[f'{platform}_user2'], label='User 2', color='pink')
    axs[i].set_ylabel('Unique Anzahl Inhalte')
    axs[i].set_ylim(0,15)
    axs[i].legend()
    axs[i].grid(True)

plt.xlabel('Datum')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

#ergebnisse
t_test_r, aligned_diversity_platform.head()
```

```python
#HYPOTHESE 3

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.formula.api as smf


file_path = '/content/Datensatz1_v2.csv'
data = pd.read_csv(file_path, delimiter=';')
data['datetime'] = pd.to_datetime(data['Date'] + ' ' + data['timestamp'], format='%d.%m.%y %H:%M')#zeit umformatieren
user1_data = data[data['User_ID'] == 1]
user2_data = data[data['User_ID'] == 2]

#berechnung shannon-e
def shannon_entropy(labels): #definiion der funktion
    counts = np.bincount(labels) #wie oft ist jedes element in counts vorhanden
    probabilities = counts[np.nonzero(counts)] / len(labels) #alle positiven werte (über null) durch die länge der labels teilen
    return -np.sum(probabilities * np.log2(probabilities)) #berechnung anhand shannen-e formel

#berechnung der entropie für jeden tag
def calc_e(data, time_column, content_column):
    data['date'] = data[time_column].dt.date #datum von zeitstempel extrahiert, keine uhrzeiten
    #berechnung der entropie für jeden tag
    #inhalte werden zu kategorischen daten umgewandelt
    daily_entropy = data.groupby('date').apply(lambda x: shannon_entropy(x[content_column].astype('category').cat.codes)).reset_index()
    daily_entropy.columns = ['date', 'entropy']
    return daily_entropy

#entropie für jeden tag für jeden user
user1de = calc_e(user1_data, 'datetime', 'content')
user2de = calc_e(user2_data, 'datetime', 'content')
ae = pd.merge(user1de, user2de, on='date', suffixes=('_user1', '_user2'))

#berechne die differenzen der entropien für jeden user für jeden tag
ae['entropy_difference'] = ae['entropy_user1'] - ae['entropy_user2']

#lineare regression für entropie unterschiede im verlauf der zeit
ae['time'] = (pd.to_datetime(ae['date']) - pd.to_datetime(ae['date'].min())).dt.days
lm = smf.ols('entropy_difference ~ time', data=ae).fit()
regression_p_val = lm.pvalues['time']

#plots, visualisierung
plt.figure(figsize=(12, 6))
plt.plot(ae['date'], ae['entropy_difference'], color='purple', marker='o', linestyle='-')
plt.axhline(0, color='black', linestyle='--')
plt.xlabel('Datum')
plt.ylabel('Entropie Differenz')
plt.xticks(rotation=45)
plt.grid(True)


for i, row in ae.iterrows():
    plt.annotate(f"{row['entropy_difference']:.2f}", (row['date'], row['entropy_difference']), textcoords="offset points", xytext=(0,10), ha='center')

plt.tight_layout()
plt.show()


#anzeige
regression_results = {
    "Linear Regression p-value for time": regression_p_val
}

regression_results

#6. und 7. tag negativ....
```

```python
#HYPOTHESE 4
import seaborn as sns

#content_name_mapping definition
content_mapping_names = {
    'Sonstige': 'Miscellaneous',
    'People and Blog': 'People and Blog',
    'Comedy and Entertainment': 'Comedy and Entertainment',
    'DIY': 'DIY',
    'Tanz-Challenge': 'Dance Challenge',
    'Food': 'Food',
    'Lipsync-Challenge': 'Lipsync Challenge',
    'Travel and Events': 'Travel and Events',
    'News and Politics': 'News and Politics',
    'Fitness and Health': 'Fitness and Health',
    'Howto and Style': 'Howto and Style',
    'Filter': 'Filter',
    'Pets and Animals': 'Pets and Animals',
    'Oddly satisfying videos': 'Oddly satisfying videos'
}

#erneutes hochladen des datensatzes
file_path_new = '/content/Datensatz1_v2.csv'
data_new = pd.read_csv(file_path_new, delimiter=';')

#funktion definieren um meiste anzeigen von inhaltskategorien anzuzeigen
def plot_top_content_frequency(user_data, user_id, platform):
    platform_fil = user_data[user_data['platform'] == platform] #daten nach plattform filtern

    #top 3 inhalte
    top_contents = platform_fil['content'].value_counts().nlargest(3).index
    top_data = platform_fil[platform_fil['content'].isin(top_contents)]

    #inhalte die am häufigsten angezeigt wurden
    top_content_freq = top_data.groupby(['Date', 'content']).size().reset_index(name='count')
    top_content_freq['content'] = top_content_freq['content'].map(content_mapping_names)

    #Diagramme 6 stück
    plt.figure(figsize=(14, 6))
    sns.lineplot(data=top_content_freq, x='Date', y='count', hue='content', marker='o')
    plt.title(f'Top 3 meist angezeigte Inhalte für User {user_id} auf {platform}')
    plt.xlabel('Datum')
    plt.ylabel('Häufigkeit der Inhalte')
    plt.ylim(0,100)
    plt.legend(title='Content', bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.show()

#Diagramme für jede einzelne Plattform
platforms = ['Youtube Shorts', 'Instagram Reels', 'Tiktok']
for platform in platforms:
    plot_top_content_frequency(data_new[data_new['User_ID'] == 1], user_id=1, platform=platform)
    plot_top_content_frequency(data_new[data_new['User_ID'] == 2], user_id=2, platform=platform)


```
